```python
cat=df["Ship Mode"].value_counts()
```

```python
cat
```

```python
explode_list=[0,0.1,0.1,0.1]
colors_list=["green","orange","blue","yellow"]
cat.plot(kind="pie",figsize=(15,5),autopct="%1.1f%%",shadow=True,startangle=180,explode=explode_list,labels=None,pctdistance=1.12,colors=colors_list)
plt.title("Ship Mode",fontsize="18")
plt.axis("equal")
plt.legend(labels=cat.index,loc="upper left")
plt.show()
```

```python
ps=df.groupby("Category")[["Profit","Sales"]].sum()
ps
```

```python
import matplotlib.pyplot as plt
ps.plot(kind='bar',figsize=(6,4))
plt.title("Profit vs Sales",fontsize=20)
plt.ylabel("Profit vs Sales")
plt.show()
```

```python
sub=df["Sub-Category"].value_counts()
sub
```

```python
x=df.groupby(["Category","Sub-Category"]).Sales.sum()
y=df.groupby(["Category","Sub-Category"]).Profit.sum()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Region",hue="Ship Mode",data=df)
plt.title("Region wise Ship Mode count")
plt.ylabel("Ship Mode count")
plt.show()
```

```python
df.groupby("Region")[["Sales","Profit"]].sum().sort_values(by=["Sales","Profit"],ascending=False).plot(kind="bar",color=["green","yellow"])
plt.ylabel("Sales and Profit")
plt.xlabel("Region")
plt.title("Sales and Profit distribution among segments")
plt.show()
```

```python
cs=df.groupby("Segment").Sales.sum()
cp=df.groupby("Segment").Profit.sum()
```

```python
ax0=cs.plot(kind="bar",figsize=(10,6),color="Purple")
ax0=cp.plot(kind="bar",figsize=(10,6),color="pink")
ax0.set_title("Category wise Sales & Profit",fontsize=18)
ax0.set_ylabel("Total slaes & profit",fontsize=18)
ax0.set_xlabel("Segment",fontsize=18)
plt.show()
```

```python
cats=df["Category"].value_counts()
cats
```
